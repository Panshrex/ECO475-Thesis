```python
import pandas as pd
import numpy as np
```

```python
from google.colab import drive, files
drive.mount('/content/drive')
```

```python
stations = pd.read_csv("/content/drive/MyDrive/University Shit /ECO475/Data/alt_fuel_stations (Jan 27 2023).csv")
stations = stations[pd.to_numeric(stations['ZIP'], errors='coerce').notnull()]
stations["ZIP"] = stations["ZIP"].astype("int64")
stations.reset_index(inplace = True, drop = True)
stations
```

```python
zip_county = pd.read_excel("/content/drive/MyDrive/University Shit /ECO475/Data/ZIP_COUNTY_122021.xlsx")
zip_county["zip"] = zip_county["zip"].astype("int64")
zip_county
```

```python
stations_merged = pd.merge(left = stations, right = zip_county, how = "inner", left_on = "ZIP", right_on = "zip")
stations_merged
```

```python
stations_merged.to_csv("stations merged.csv")
files.download("stations merged.csv")
```
